```python
import pandas as pd 
import seaborn as sns
import numpy as np
```

```python
avo = pd.read_csv('avocado.csv')
avo.head()
```

```python
avo.info()
```

```python
date_cols = ['Date']
avo[date_cols].apply(pd.to_datetime)
```

```python
avo.info()
```

```python
avo.query('year == 2015')[['Date','Total Bags','Small Bags']]
```

```python
avomelt = avo.melt( 
id_vars = 'Date',
value_vars=['Total Bags','Small Bags'], 
var_name ='Bags', value_name='Count')
avomelt
```

```python
sns.relplot(data=avomelt, 
kind='line', x='Date', y='Count', hue='Bags')
```

```python
avo['Sales Volume'] = pd.qcut(avo['Total Volume'], q=4,
labels=['Poor','Modest','Good','Excellent'])

avo.head()
```

```python
sns.countplot(data=avo, x='year', hue='Sales Volume')
```

```python
avosize = avo.melt( 
id_vars = 'year',
value_vars=['Small Bags','Large Bags','XLarge Bags'], 
var_name ='Bags', value_name='Count')
avosize
```

```python
sns.relplot(data=avosize, kind='line', x='year', y='Count', hue='Bags')
```
